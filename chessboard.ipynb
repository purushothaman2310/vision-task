```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# reads image 'opencv-logo.png' as grayscale
img = cv2.imread(r'C:\Users\Saravanan\Downloads\edge_detection\data\Image_3.png') 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)# Converting the image to grayscale.
gray_filtered = cv2.bilateralFilter(gray, 10, 100, 100)# Smoothing without removing edges.
edges_filtered = cv2.Canny(gray_filtered, 80, 160)## Applying the canny filter
edges = cv2.Canny(gray,100,200)#The Canny filter thresholds can be tuned to catch only the strongest edges and get cleaner contours. The higher the thresholds, the cleaner the edges.

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges_filtered,cmap = 'Greens')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()


```

```python
# filteration Method gives  good accuracy and results
import numpy as np
import cv2
import matplotlib.pyplot as plt

# reads image from path file
img = cv2.imread(r'C:\Users\Saravanan\Downloads\edge_detection\data\Image_2.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)# Converting the image to grayscale.
gray_filtered = cv2.bilateralFilter(gray, 10, 100, 100)# Smoothing without removing edges.
edges_filtered = cv2.Canny(gray_filtered, 80, 160)#Applying the canny filter with Hystersis values

plt.subplot(121),plt.imshow(img,cmap = 'gray')#for Locating the image in plot
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges_filtered,cmap = 'Greens')# result image with green lines
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()

# I didn't uses functions and don't want to make program complex
# My aim is  beginner in python should understand my code
# Done by purushothaman saravanan
```
